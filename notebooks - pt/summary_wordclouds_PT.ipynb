# CRIAR NUVEM DE PALAVAS DE COLEÇÕES DE CURRÍCULOS LATTES

O código neste Notebook Jupyter extrai o resumo dos Currículos Lattes de uma determinada lista de pastas e, com as palavras mais bem classificadas - de acordo com o algoritmo TF-IDF, ele gera uma nuvem de palavras.

Explicações para o código seguirão cada célula. A célula abaixo é reservada para definições e carregamento de pacotes Python.

```python
import LattesLab as ll
import os
```

A variável "folderlist" contém o nome das pastas onde a coleção de Currículos Lattes é encontrada.

A variável workdir armazena a pasta de trabalho atual do aplicativo, onde os arquivos podem ser salvos, se desejado.

```python
folderlist = ["D:\\thiag\\Documents\\INPE\\Research\\Datasets\\PIBIC"]
workdir = os.getcwd()
```

A função ll.get_lattes_desc_folders retorna a lista de resumos de cada Currículo Lattes da lista de pastas fornecida. Se nenhum resumo for encontrado para um dado currículo Lattes, ele será considerado como um resumo vazio.

```python
summarylist = ll.get_lattes_desc_folders(folderlist)
```

Para confirmar como o código funciona, uma lista dos cinco primeiros resumos é solicitada pelo comando na célula abaixo.

```python
summarylist[:5]
```

A função ll.summary_list_top_words executa o algoritmo TF-IDF, usando cada resumo como um documento separado. É possível definir palavras a serem descartadas da análise, se necessário. As 50 principais palavras e suas pontuações são classificadas por pontuação em uma nova variável chamada "topwords".

```python
topwords = ll.summary_list_top_words(summarylist, 50)
```

Vamos verificar o conteúdo das topwords variáveis.

```python
topwords
```

O vetor é então transformado para ser usado como uma entrada para o comando WordCloud().generate().

Finalmente, a núvem de palavras pode ser exibido.

```python
ll.word_list_to_cloud(workdir, topwords)
```
