# Termos Mais Frequentes nos Títulos das Publicações

O código deste Notebook Jupyter extrai os dados dos Currículos Lattes que se encontram em uma lista de pastas e cria nuvens de palavras dos termos mais frequentes encontrados nos títulos das publicações de cada pesquisador.

Explicações para o código seguirão cada célula. A célula abaixo é reservada para definições e carregamento de pacotes Python.

```python
import LattesLab as ll
```

A variável "folderlist" contém a lista de caminhos de pastas onde a coleção de Currículos Lattes é encontrada.

```python
folderlist = ["C:\\Users\\thiag\\Desktop\\work"]
```

A função get_pub_dataframe_from_folders cria uma matriz de dados (dataframe) a partir dos Crrículos Lattes nas pastas. O segundo parâmetro (True ou False) define se esse dataframe será salvo em um arquivo csv para uso posterior.

```python
pubframe = ll.get_pub_dataframe_from_folders(folderlist, False)
```

Para ver como é o dataframe, vamos dar uma olhada nas primeiras linhas de seus dados.

```python
pubframe.head()
```

A função get_wordclouds_from_pub_frame cria núvens de palavras com os termos mais frequentes encontrados nos títulos das publicações de cada pesquisador, salvando os resultados nas seguintes variáveis:

- a é a lista dos nomes dos pesquisadores;
- b é a lista da lista de títulos publicados por cada pesquisador;
- c é a lista dos termos mais frequentes encontrados nos títulos das publicações do pesquisador.

O primeiro parâmetro da função é o dataframe recuperado da função get_pub_dataframe_from_folders. O segundo parâmetro (True ou False) define se esse dataframe será salvo em um arquivo csv para uso posterior.

```python
[a, b, c] = ll.get_wordclouds_from_pub_frame(pubframe, False)
```
