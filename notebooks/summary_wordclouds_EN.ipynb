# CREATE WORDCLOUD FROM LATTES CV COLLECTIONS

The code in this notebook extracts the summary of the Lattes CVs of a given list of folders and, with the highest ranking words - according to TF-IDF algorithm, it generates a word cloud.

Explanations for the code will follow each cell. The cell below is reserved for definitions and load of Python packages.

```python
import LattesLab as ll
import os
```

The variable "folderlist" contains the name of the folder where the Lattes CV collection is found.

The variable workdir stores the current working folder of the application, where files can be saved, if desired.

```python
folderlist=["C:\\Users\\thiag\\Desktop\\work\\"]
workdir = os.getcwd()
```

The function ll.get_lattes_desc_folders returns the list of summaries of each Lattes CVs from the given list of folders. If no summary is found for a given Lattes CV, it is assumed as an empty summary.

```python
summarylist = ll.get_lattes_desc_folders(folderlist)
```

To verify that the code has worked, a list of the five first summaries is requested by the command in the cell below.

```python
summarylist[:5]
```

The function ll.summary_list_top_words runs the TF-IDF algorithm, using each summary as a separate document. The stop-words may be changed, if required. The top 50 words and its scores are sorted by score on a new variable named "topwords".

```python
topwords = ll.summary_list_top_words(summarylist, 50)
```

Let's verify the contents of the variable topwords.

```python
topwords
```

The vector is then transformed to be used as an input to the WordCloud().generate() command.

Finally, the wordcloud may be exhibited.

```python
ll.word_list_to_cloud(workdir, topwords)
```
