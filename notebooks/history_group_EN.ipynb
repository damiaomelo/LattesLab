# Finding the Production History of a Researcher's Group

The LattesLab library may be used to retrieve the productivity results of a group of researchers. Once the Lattes CVs of these researchers are downloaded and stored in a folder, the analysis may be conducted, according to the steps of this jupyter Notebook.

First of all, the LattesLab library must be imported.

```python
import LattesLab as ll
```

In the following cells, the variables are defined:
- "folder" is a string with the folder that holds the Lattes CV files of the group of researchers
- "params" is a list of strings. Each string of "params" is a parameter to be retrieved from the Lattes CV file. Must be a subset of 'papers', 'works', 'orientations', 'chapters' and 'books'
- "nyears" is an integer containing the number of years, from the current year backwards, to be retrieved from the Lattes CVs.

```python
folder = 'C:\\Users\\thiag\\Desktop\\work\\'
params = ['papers', 'works', 'orientations', 'chapters', 'books']
nyears = 20
```

Finally, the history can be discovered with the LattesLab function get_history_file_n.

```python
myhist = ll.get_history_file_n(folder, params, nyears)
```
